## Importando librerías

```python
from keras.datasets import mnist
from keras import layers, models
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

## Nuestro set de datos

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creando un modelo de datos

```python
model = models.Sequential()
model.add(layers.Dense(512,activation='relu', input_shape=(28*28,)))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Limpieza de datos

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255

x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
```

```python
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
train_labels[85]
```

```python
y_train[85]
```

## Entrenando el modelo

```python
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

## Evaluando sobre data de test

```python
model.evaluate(x_test,y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from google.colab import files
import io

# 1. Subir los archivos (Selecciona tus 9 imágenes al mismo tiempo)
print("📤 Selecciona tus 9 imágenes (puedes seleccionarlas todas juntas):")
uploaded = files.upload()

def procesar_limpieza_colab(contenido_archivo):
    # Abrir la imagen desde los datos subidos
    img = Image.open(io.BytesIO(contenido_archivo)).convert('L')

    # Invertir colores: Papel blanco -> Fondo negro
    img = ImageOps.invert(img)

    # --- LIMPIEZA EXTREMA ---
    # Esto elimina el fondo gris para que el modelo no se confunda con el '3'
    img_np = np.array(img)
    umbral_fuerte = 150 # Si el número desaparece, baja este valor a 120
    img_np = (img_np > umbral_fuerte) * 255

    # Volver a imagen y redimensionar a 28x28
    img_limpia = Image.fromarray(img_np.astype('uint8'))
    img_28x28 = img_limpia.resize((28, 28), Image.Resampling.LANCZOS)

    # Normalizar y aplanar (784 neuronas de entrada)
    img_final = np.array(img_28x28).astype('float32') / 255.0
    img_vector = img_final.reshape(1, 784)

    return img_vector, img_28x28

# 2. Procesar cada archivo subido por separado
if not uploaded:
    print("❌ No subiste ningún archivo.")
else:
    print(f"\n✅ Procesando {len(uploaded)} archivos individualmente...\n")

    for nombre_archivo, contenido in uploaded.items():
        try:
            # Procesar
            datos, img_vista = procesar_limpieza_colab(contenido)

            # Predicción con tu modelo entrenado
            prediccion = model.predict(datos, verbose=0)
            resultado = np.argmax(prediccion)
            confianza = np.max(prediccion) * 100

            # --- MOSTRAR RESULTADOS INDIVIDUALES ---
            print(f"📄 ARCHIVO: {nombre_archivo}")
            print(f"🤖 PREDICCIÓN: {resultado} | CONFIANZA: {confianza:.2f}%")

            plt.figure(figsize=(2, 2))
            plt.imshow(img_vista, cmap='gray')
            plt.axis('off')
            plt.show()
            print("-" * 40)

        except Exception as e:
            print(f"❌ Error al procesar {nombre_archivo}: {e}")
```
