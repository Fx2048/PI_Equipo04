## Importando librerías

```python
from keras.datasets import mnist
from keras import layers, models
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

## Nuestro set de datos

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creando un modelo de datos

```python
model = models.Sequential()
model.add(layers.Dense(512,activation='relu', input_shape=(28*28,)))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Limpieza de datos

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255

x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
```

```python
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
train_labels[85]
```

```python
y_train[85]
```

## Entrenando el modelo

```python
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

## Evaluando sobre data de test

```python
model.evaluate(x_test,y_test)
```

```python
from google.colab import files
from PIL import Image, ImageOps, ImageFilter
import numpy as np
import matplotlib.pyplot as plt

# 1. Subir imagen
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 2. Cargar y convertir a escala de grises
img = Image.open(filename).convert('L')

# 3. Auto-Inversión: MNIST necesita trazo blanco sobre fondo negro
# Si el promedio es > 127, el fondo es claro y debemos invertir
img_data = np.array(img)
if np.mean(img_data) > 127:
    img = ImageOps.invert(img)

# 4. Limpieza y Engrosado
# Eliminamos grises suaves (sombras) y dejamos solo el trazo fuerte
img = img.point(lambda p: 255 if p > 150 else 0)
# Engrosamos un poco el trazo para que la red lo vea mejor
img = img.filter(ImageFilter.MaxFilter(3))

# 5. Redimensionar a 28x28 (tamaño MNIST)
img = img.resize((28, 28), Image.LANCZOS)

img_array = np.array(img)

# 6. Normalizar y Aplanar
img_array = img_array.astype('float32') / 255
img_ready = img_array.reshape(1, 28*28)

# 7. Visualizar lo que la IA procesa
plt.imshow(img_array, cmap='gray')
plt.title("Entrada final al modelo")
plt.show()

# 8. Predicción
prediction = model.predict(img_ready)
numero = np.argmax(prediction)
confianza = np.max(prediction) * 100

print(f"\n>>> NÚMERO PREDICHO: {numero} <<<")
print(f"Confianza: {confianza:.2f}%")
```
