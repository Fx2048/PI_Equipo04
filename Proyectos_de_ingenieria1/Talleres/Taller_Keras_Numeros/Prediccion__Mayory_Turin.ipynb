## Importando librerías

```python
from keras.datasets import mnist
from keras import layers, models
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

## Nuestro set de datos

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creando un modelo de datos

```python
model = models.Sequential()
model.add(layers.Dense(512,activation='relu', input_shape=(28*28,)))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Limpieza de datos

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255

x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
```

```python
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
train_labels[85]
```

```python
y_train[85]
```

## Entrenando el modelo

```python
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

## Evaluando sobre data de test

```python
model.evaluate(x_test,y_test)
```

```python
# IMPORTANTE: librerías adicionales
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# 1. Subir imagen desde tu computadora
uploaded = files.upload()

# 2. Obtener el nombre del archivo
image_path = list(uploaded.keys())[0]

# 3. Abrir imagen
img = Image.open(image_path)

# 4. Convertir a escala de grises (porque MNIST es grayscale)
img = img.convert('L')

# 5. Redimensionar a 28x28 píxeles
img = img.resize((28, 28))

# 6. Convertir a array
img_array = np.array(img)

# 7. Invertir colores (IMPORTANTE)
# MNIST: fondo negro, número blanco
img_array = 255 - img_array

# 8. Normalizar (igual que entrenamiento)
img_array = img_array.astype('float32') / 255

# 9. Aplanar (reshape a 784)
img_array = img_array.reshape(1, 28*28)

# 10. Mostrar imagen procesada
plt.imshow(img_array.reshape(28,28), cmap='gray')
plt.title("Imagen procesada")
plt.axis('off')
plt.show()

# 11. Predicción
prediction = model.predict(img_array)

# 12. Obtener número final
predicted_digit = np.argmax(prediction)

print("Número predicho:", predicted_digit)
```
